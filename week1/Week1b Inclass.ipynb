```python
import numpy as np
import matplotlib.pyplot as mpl
```

```python
def datagen(n):
    c=np.random.randint(1,3,size=(n))
    us=np.random.normal(loc=0,scale=1,size=(2,n))

  
    #######################
    # your code here
    a_pre = np.array([[np.cos(np.pi/4), np.sin(np.pi/4)],
                      [-np.sin(np.pi/4), np.cos(np.pi/4)]])
    a_post = np.array([[ 3, 0],
                       [ 0, 1]])

    mu1 = 0
    mu2 = 2.5

    A = np.matmul(a_pre, a_post)
    xs = np.matmul(A,us)
    ys = np.zeros(n) 

    for i in range(n):
        if c[i] == 1: 
            xs[:,i] += mu1
            ys[i] = int(np.random.random() > 0.3)

        elif c[i] == 2:
            xs[:,i] += mu2
            ys[i] = int(np.random.random() > 0.6)

    #######################
#     print('xs:',xs,'\nyx', ys,'\nc', c)
    return xs,ys,c
```

```python
def plot1(n):
    xs,ys,c=datagen(n)

    #######################
    # your code here
    inds = np.where(ys==0)
    mpl.plot(xs[0,inds[0]],xs[1,inds[0]],'r.')
    
    inds = np.where(ys==1)
    mpl.plot(xs[0,inds[0]],xs[1,inds[0]],'b.')
    mpl.show(block=True)
    #######################
```
